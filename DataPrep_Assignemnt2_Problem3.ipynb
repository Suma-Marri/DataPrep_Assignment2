```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/DPFONAL_1.csv")
df = df.dropna(axis= 0, how='any', subset=['marital'])
df.head()
```

```python
df.groupby('marital').mean()
```

```python
df['marital'].replace(['U', 'M', 'S', 'B', 'A'], [1, 2, 3, 4, 5], inplace=True)
```

```python
from statsmodels.formula.api import ols
from patsy.contrasts import Treatment
from patsy.contrasts import Diff
```

```python
levels = [1,2,3,4,5]
```

```python
pd.get_dummies(df.marital.values, drop_first=False)
```

```python
contrast = Diff().code_without_intercept(levels)
```

```python
print(contrast.matrix)
```

```python
mod = ols("rev_Mean ~ C(marital, Diff)", data=df)
res = mod.fit()
print(res.summary())
```
